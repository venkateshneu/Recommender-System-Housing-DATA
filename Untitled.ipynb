```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Sample data with missing values
data_categorical = {
    'Education': ['Bachelors', 'Masters', 'High School', np.nan, 'PhD', 'High School', 'Masters', 'Bachelors']
}

df_categorical = pd.DataFrame(data_categorical)

# Encode categorical data
encoder = OneHotEncoder(handle_unknown='ignore')
df_encoded = pd.DataFrame(encoder.fit_transform(df_categorical[['Education']]).toarray(), columns=encoder.get_feature_names_out(['Education']))

```

```python
df_encoded
```

```python
# Initialize KNNImputer
knn_imputer_categorical = KNNImputer(n_neighbors=2)

# Fit and transform the encoded data
imputed_data_categorical = knn_imputer_categorical.fit_transform(df_encoded)

# Convert array back to DataFrame with original columns
df_imputed_categorical = pd.DataFrame(imputed_data_categorical, columns=df_encoded.columns)

print("Imputed Categorical Data:")
print(df_imputed_categorical)
```
